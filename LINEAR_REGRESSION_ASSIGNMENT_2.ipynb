**Importing The Necessary Libraries and Read the Data set**

```python
import pandas as pd
somoire = pd.read_csv('/content/SomoireDataset.csv')
somoire
```

**Describe the Data Set**

```python
somoire.describe()
```

**Importing the Seaborn Libary and Visualize**

```python
import seaborn as sns
sns.heatmap(somoire)
```


**Identifying the Dependent Variables and the Independent Variables and using the same seaborn library us above to visualize**



```python
x = somoire.drop('Accident Outcome', axis = 1)
y = somoire['Accident Outcome']

import seaborn as sns
sns.heatmap(somoire)
```

**Importing sklearn library to split data into train and test data**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=0)
```

**importing LinearRegression Model to train the data**

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
prediction = model.predict(x_test)
prediction
```

```python
from sklearn.metrics import mean_absolute_error
accuracy = mean_absolute_error(y_test,prediction)
accuracy
```

```python
import joblib
joblib.dump(model,'Accident Predictor')
```
